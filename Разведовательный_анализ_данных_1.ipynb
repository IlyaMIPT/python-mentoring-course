```python
import pandas as pd

df = pd.read_csv('students_performance.csv')
import matplotlib.pyplot as plt
plt.close("all")
plt.style.use('ggplot')
%matplotlib inline
df.head(10)
```

```python
columns_new_name = ['gender', 'ethnicity', 
                    'parental_level_of_education',
                    'lunch',	
                    'test_preparation_course',	
                    'math_score',
                    'reading_score',
                    'writing_score']
df.columns = columns_new_name
df.head(3)
```

            Считаем сумму баллов, среднее по каждому студенту и данным

```python
df['sum_score'] = df[['math_score', 'reading_score', 'writing_score']].sum(axis='columns')
df['mean_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis='columns').round(2)

df.head()
```

```python
df.describe().round(3)
```

```python
median_and_var = df[['math_score', 'reading_score', 'writing_score','sum_score', 'mean_score']].agg(['mean', 'median','var', 'std'])
median_and_var.round(2)
```

```python
df['parental_level_of_education'].value_counts()
```

```python
df
```

            Зависимость оценок от гендера

```python
t = df.groupby('gender')[['gender', 'sum_score', 'mean_score']].agg(count=('gender', 'count'),
    mean_sum_score=('sum_score', 'mean'),
    mean_score=('mean_score', 'mean')).round(3)

t
```

```python
mean_and_sum_for_gender = t[['mean_sum_score', 'mean_score']]


ax = mean_and_sum_for_gender.plot.bar(title='Сумма баллов и среднее для гендера')
                               
ax.set_ylim(bottom=0, top=225)
plt.show()
```

            Общая зависимость оценок от гендора и уровня образования родителей

```python
t = df.groupby(by=['parental_level_of_education', 'gender']).agg(count=('gender', 'count'),
    mean_sum_score=('sum_score', 'mean'),
    mean_score=('mean_score', 'mean'))
t = t.unstack()
t.columns = ['_'.join(col).strip() for col in t.columns.values]
t.round(2)
```

```python
t = t.drop(['mean_sum_score_female', 'mean_sum_score_male'],axis=1)
t.plot.bar( figsize=(16,8), title='Количество студентов разного пола и их средний балл в зависимости от образования родителей',
                                ylabel = 'Показатель',
                                xlabel = 'Уровень образования родителей').set_ylim(bottom=0, top=150)

plt.show()
```

            Зависимость оценок по предметам от пола и уровня образования родителей

```python
t = df.groupby(by=['parental_level_of_education', 'gender']).agg(
    mean_math_score=('math_score', 'mean'),
    mean_reading_score=('reading_score', 'mean'), 
    mean_writing_score=('writing_score', 'mean'),                                                                                                               
    )
t = t.unstack()
t.columns = ['_'.join(col).strip() for col in t.columns.values]
t.round(2)
```

```python
t.plot.bar( figsize=(16,8), title='Средний балл по предметам в зависимости гендера и образования родителей',
                                ylabel = 'Показатель',
                                xlabel = 'Уровень образования родителей').set_ylim(bottom=0, top=100)

plt.show()
```
